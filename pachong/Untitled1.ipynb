```python
with open('python_zhaoping.csv','w',encoding='utf-8') as f:
    url = 'https://sou.zhaopin.com/jobs/searchresult.ashx?jl=%E5%8E%A6%E9%97%A8&kw=python&sm=0&p=1'
    res = requests.get(url).text
    p = etree.HTML(res)
    
    job = p.xpath('//*[@id="newlist_list_content_table"]/table[1]/tr/th[1]/span/text()')[0]
    company = p.xpath('//*[@id="newlist_list_content_table"]/table[1]/tr/th[3]/text()')[0]
    money = p.xpath('//*[@id="newlist_list_content_table"]/table[1]/tr/th[4]/text()')[0]
    address = p.xpath('//*[@id="newlist_list_content_table"]/table[1]/tr/th[5]/text()')[0]
    date = p.xpath('//*[@id="newlist_list_content_table"]/table[1]/tr/th[6]/text()')[0]
    f.write('{},{},{},{},{}\n'.format(job,company,money,address,date))
    for i in range(7):
        url = 'https://sou.zhaopin.com/jobs/searchresult.ashx?jl=%E5%8E%A6%E9%97%A8&kw=python&sm=0&p={}'.format(i+1)
        res = requests.get(url).text
        p = etree.HTML(res)
        file =p.xpath( '//*[@id="newlist_list_content_table"]/table')
        time.sleep(3)
        for div in file:
            n1 = div.xpath('./tr[1]/td[1]/div/a//text()')
            job = ''.join(n1)
            company = div.xpath('./tr[1]/td[3]/a[1]//text()')
            if len(company)>0:
                company = company[0]
            else:
                company = None
                
            money = div.xpath('./tr[1]/td[4]//text()')
            if len(money)>0:
                money = money[0]
            else:
                money = None
            address = div.xpath('./tr[1]/td[5]//text()')
            if len(address)>0:
                address =address[0]
            else:
                address =None
            date = div.xpath('./tr[1]/td[6]/span//text()')
            if len(date)>0:
                date = date[0]
            else:
                date = None
            f.write('{},{},{},{},{}\n'.format(job,company,money,address,date))
            
            
            
```
