```python
import requests
from lxml import etree
```

```python
url = 'https://movie.douban.com/subject/1292052/'

data = requests.get(url).text
data
```

```python
s = etree.HTML(data)
s
```

```python
film = s.xpath('//*[@id="content"]/h1/span[1]/text()')
print(film)
```

```python
director = s.xpath('//*[@id="info"]/span[1]/span[2]/a/text()')
actor1 = s.xpath('//*[@id="info"]/span[3]/span[2]/span[1]/a/text()')
actor2 = s.xpath('//*[@id="info"]/span[3]/span[2]/span[2]/a/text()')
actor3 = s.xpath('//*[@id="info"]/span[3]/span[2]/span[3]/a/text()')
releasedate = s.xpath('//*[@id="info"]/span[10]/text()')
time = s.xpath('//*[@id="info"]/span[13]/text()')
```

```python
actor= s.xpath('//*[@id="info"]/span[3]/span[2]')
```

```python
print('电影名称：',film)
print('导演：',director)
print('主演：',actor)
print('上映时间：',releasedate)
print('片长：',time)
```
